```python
# Base de datos temporal (Lista de diccionarios)
enfermeros = []

def agregarenfermero():
    print("\n--- Registrar Nuevo Enfermero ---")
    id_enf = input("ID o Cédula: ")
    nombre = input("Nombre completo: ")
    especialidad = input("Especialidad: ")

    enfermero = {
        "id": id_enf,
        "nombre": nombre,
        "especialidad": especialidad
    }
    enfermeros.append(enfermero)
    print(f"¡Enfermero {nombre} agregado con éxito!")

def listarenfermero():
    print("\n--- Lista de Enfermeros ---")
    if not enfermeros:
        print("No hay enfermeros registrados.")
    else:
        for i, e in enumerate(enfermeros, 1):
            print(f"{i}. ID: {e['id']} | Nombre: {e['nombre']} | Especialidad: {e['especialidad']}")

def buscarenfermero():
    print("\n--- Buscar Enfermero ---")
    busqueda = input("Ingrese el ID o Nombre a buscar: ").lower()
    encontrado = False
    for e in enfermeros:
        if busqueda == e['id'].lower() or busqueda in e['nombre'].lower():
            print(f"Resultado: ID: {e['id']}, Nombre: {e['nombre']}, Especialidad: {e['especialidad']}")
            encontrado = True
    if not encontrado:
        print("No se encontró ningún enfermero con ese criterio.")

def modificarenfermero():
    print("\n--- Modificar Datos ---")
    id_mod = input("Ingrese el ID del enfermero a modificar: ")
    for e in enfermeros:
        if e['id'] == id_mod:
            print(f"Editando a: {e['nombre']}")
            e['nombre'] = input("Nuevo nombre (deje vacío para no cambiar): ") or e['nombre']
            e['especialidad'] = input("Nueva especialidad (deje vacío para no cambiar): ") or e['especialidad']
            print("Datos actualizados correctamente.")
            return
    print("ID no encontrado.")

def borrarenfermero():
    print("\n--- Borrar Enfermero ---")
    id_del = input("Ingrese el ID del enfermero a eliminar: ")
    for i, e in enumerate(enfermeros):
        if e['id'] == id_del:
            eliminado = enfermeros.pop(i)
            print(f"Enfermero {eliminado['nombre']} eliminado.")
            return
    print("No se encontró el ID.")

# Función principal para controlar el flujo con un ciclo while
def menu():
    continuar = True
    while continuar:
        print("\n==============================")
        print("   SISTEMA DE GESTIÓN (HOSPITAL)")
        print("==============================")
        print("1. Agregar Enfermero")
        print("2. Listar Enfermeros")
        print("3. Buscar Enfermero")
        print("4. Modificar Enfermero")
        print("5. Borrar Enfermero")
        print("6. Salir")

        opcion = input("\nSeleccione una opción (1-6): ")

        if opcion == "1":
            agregarenfermero()
        elif opcion == "2":
            listarenfermero()
        elif opcion == "3":
            buscarenfermero()
        elif opcion == "4":
            modificarenfermero()
        elif opcion == "5":
            borrarenfermero()
        elif opcion == "6":
            print("Saliendo del sistema... ¡Buen día!")
            continuar = False
        else:
            print("Opción no válida, intente de nuevo.")

# Ejecutar el programa
menu()
```
